```python
from abc import ABC, abstractmethod
```

```python
# use abstract classes to define the structure (i.e., methods)
# that need to be in the class, and then inherit it when making
# your actual classes
class BaseAPI(ABC):
    @abstractmethod
    def __init__(self, name:str):
        pass

    @abstractmethod
    def get_data(self, n: int) -> list:
        pass
```

```python
class RealAPI(BaseAPI):
    def __init__(self, name:str):
        self.name = name
        return
```

```python
x = RealAPI()
```

```python
class RealAPI:
    def __init__(self, name:str):
        self.name =  name
    def get_data(self, n:int) -> list:
        return list(range(n))

class DummyAPI(RealAPI):
    def __init__(self, name:str):
        super().__init__(name)
    def get_data(self, *args, **kwargs):
        return 1

class MyProgram:
    def __init__(self,
                 use_api:bool = True
                ):
        if use_api:
            self.api = RealAPI(name='Real API')
        else:
            self.api = DummyAPI(name='Fake API')

    def get_data(self, n:int):
        return self.api.get_data(n)
```
